```python
import subprocess
from os.path import join 

class PlinkConverter:
    def __init__(self, input_file, output_prefix, output_path):
        self.input_file = input_file
        self.output_prefix = output_prefix
        self.output_path = output_path
        self.plink_path = r"C:\Users\greys\OneDrive\Downloads\plink2\plink2.exe"
        
    def convert_to_ped_map(self):
        # Read in the input file and convert to PED and MAP formats
        # Replace this with your own code to parse the input file and create PED and MAP files

        # Example code to read in sample and genotype data from a tab-separated file
        sample_data = {}
        genotype_data = {}
        with open(self.input_file, "r") as input_file:
            header = input_file.readline().strip().split("\t")
            for line in input_file:
                fields = line.strip().split("\t")
                sample_id = fields[0]
                phenotype = fields[1]
                sample_data[sample_id] = phenotype
                genotypes = fields[2:]
                genotype_data[sample_id] = genotypes

        ped_file = join(self.output_path, f"{self.output_prefix}.ped")
        map_file = join(self.output_path, f"{self.output_prefix}.map")
        # Create the PED file
        with open(ped_file, "w") as ped_out:
            # Write out the sample IDs and phenotype values
            for sample_id, phenotype in sample_data.items():
                ped_out.write(
                    f"{sample_id}\t{sample_id}\t0\t0\t0\t{phenotype}\t")
                # Write out the genotype data
                for genotype in genotype_data[sample_id]:
                    ped_out.write(f"{genotype}\t")
                ped_out.write("\n")
        # Create the MAP file
        with open(map_file, "w") as map_out:
            for i, snp_id in enumerate(header[2:], start=1):
                map_out.write(f"0\t{snp_id}\t0\t{i}\n")
                
        print(f'PED and MAP files created in: {self.output_path=}')
        
    def run_make_bed(self):
        # Call the plink2 executable to convert the PED and MAP files to binary PLINK format
        subprocess.run([self.plink_path, "ped",
                   f"{self.output_prefix}.ped", f"{self.output_prefix}.map", "--make-bed", f"--out={self.output_prefix}.bed"])

        # subprocess.run([self.plink_path, "--make-bed",
        #                f"--file={self.output_prefix}", f"--out={self.output_prefix}"])
        
        print(f'--make-bed completed in: {self.output_prefix=}')
        
converter = PlinkConverter(
    r"C:\Users\greys\projects\cosnetix_pipeline\external\genome\genomicSample.txt", 
    r"genomic-demo",
    r"C:\Users\greys\projects\cosnetix_pipeline\pipeline\gwas\data\interim")
converter.convert_to_ped_map()
converter.run_make_bed()
```

```python
import pandas as pd
from pandas_plink import write_plink


def convert_ped_map_to_bed(input_ped, input_map, output_prefix):
    # Read .ped file
    ped_data = pd.read_csv(input_ped, sep=" ", header=None)
    fam_data = pd.DataFrame(ped_data.iloc[:, :6])
    fam_data.columns = ["fid", "iid", "father", "mother", "sex", "pheno"]

    # Read .map file
    map_data = pd.read_csv(input_map, sep="\t", header=None, names=[
                           "chrom", "snp", "cm", "pos"])
    bim_data = map_data[["chrom", "snp", "cm", "pos", "pos", "pos"]].copy()
    bim_data.columns = ["chrom", "snp", "cm", "pos", "a1", "a2"]

    # Extract genotype data from .ped file
    genotype_data = ped_data.iloc[:, 6:].values.reshape(-1, len(map_data), 2)

    # Write .bed, .bim, and .fam files
    write_plink(output_prefix, (fam_data, bim_data,
                genotype_data), verbose=True)


# Example usage
input_ped = "path/to/your/input.ped"
input_map = "path/to/your/input.map"
output_prefix = "path/to/your/output_prefix"

convert_ped_map_to_bed(input_ped, input_map, output_prefix)
```

```python
import os
import subprocess
from dotenv import load_dotenv
load_dotenv()

from pipeline.data.helpers import read_yaml

class GenotypingPreprocessor:
    def __init__(self, data_contract_path:str, input_file:str):
        
        self.data_contract = read_yaml(data_contract_path)
        
        self.input_file = input_file
        self.output_prefix = self.data_contract['Genome']['output']
        # self.output_path = self.data_contract['Genome']['collections']['interim']
        self.output_path = r'C:\Users\greys\projects\cosnetix_pipeline\pipeline\gwas\data\interim'
        self.ped_file = os.path.join(self.output_path,f"{self.output_prefix}.ped")
        self.map_file = os.path.join(self.output_path,f"{self.output_prefix}.map")
        self.bed_file = os.path.join(self.output_path,f"{self.output_prefix}.bed")
        self.bim_file = os.path.join(self.output_path,f"{self.output_prefix}.bim")
        self.fam_file = os.path.join(self.output_path,f"{self.output_prefix}.fam")
        self.output_file = os.path.join(self.output_path,f"{self.output_prefix}.bed")
        self.cleaned_prefix = f"{self.output_prefix}_cleaned"
        self.vcf_file = f"{self.cleaned_prefix}.vcf"
        self.plink_path = os.getenv("PLINK_PATH")
        # self.plink_path = 'plink'
        print(f'{self.plink_path=}')
    def _check_input_file(self):
        if not os.path.isfile(self.input_file):
            print(f"Error: Input file '{self.input_file}' not found.")
            return False
        return True

    def _check_plink2(self):
        try:
            subprocess.run([self.plink_path, "--version"], check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            return True
        except FileNotFoundError:
            print("Error: plink2 not found. Please install plink2 and ensure it's in your PATH.")
            return False

    def convert_to_ped_map(self):
        print('conversion to ped and map')
        if not self._check_input_file():
            return

        with open(self.input_file, 'r') as infile, open(self.ped_file, 'w') as ped, open(self.map_file, 'w') as map:
            for line in infile:
                # Ignore comment lines
                if line.startswith("#"):
                    continue

                rsid, chromosome, position, genotype = line.strip().split('\t')
                # Write .map file
                map.write(f"{chromosome} {rsid} 0 {position}\n")

                # Write .ped file
                ped.write(f"{rsid} {chromosome} {position} {genotype}\n")

        print(f"Conversion complete. Output files are '{self.ped_file}' and '{self.map_file}'.")

    def preprocess_with_plink2(self):
        self.convert_to_ped_map()
        if not (self._check_input_file() and self._check_plink2()):
            return

        # Convert .ped and .map files to .bed, .bim, and .fam files using plink2
        plink2_preprocess_command = [
            self.plink_path,
            "--file", self.ped_file,
            "--make-bed",
            "--out", self.output_file
        ]
        try:
            subprocess.run(plink2_preprocess_command, check=True)
            print(f"Preprocessing complete. Output files are '{self.bed_file}', '{self.bim_file}', and '{self.fam_file}'.")
        except subprocess.CalledProcessError:
            print("Error: plink2 failed to preprocess the genome data.")
            return

        # Quality control and filtering
        plink2_filter_command = [
            self.plink_path,
            "--bfile", self.output_prefix,
            "--maf", "0.01",
            "--geno", "0.05",
            "--hwe", "1e-6",
            "--mind", "0.1",
            "--make-bed",
            "--out", self.cleaned_prefix
        ]
        try:
            subprocess.run(plink2_filter_command, check=True)
            print(f"Quality control and filtering complete. Output files are '{self.cleaned_prefix}.bed', '{self.cleaned_prefix}.bim', and '{self.cleaned_prefix}.fam'.")
        except subprocess.CalledProcessError:
            print("Error")


prc = GenotypingPreprocessor(
    r'C:\Users\greys\projects\cosnetix_pipeline\data-contract.yml',
    r'C:\Users\greys\projects\cosnetix_pipeline\external\genome\genomicSample.txt'
)
prc._check_plink2()
prc.preprocess_with_plink2()
```
